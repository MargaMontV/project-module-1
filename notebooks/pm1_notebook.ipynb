```python
# Import libraries
import pandas as pd
import requests    

# Pandas display options
pd.set_option('display.max_columns', None)
```

```python
end_point = 'https://datos.madrid.es/egob/' # Main End-Point
body = 'catalogo/202311-0-colegios-publicos.json'# Body
pulls_response = requests.get(end_point + body)
```

```python
status = pulls_response.status_code
status
```

```python
#content = pulls_response.content
#content
```

```python
pulls_json = pulls_response.json()
print(type(pulls_json))
print(len(pulls_json))
#pulls_json
```

```python
df_col_pub = pd.json_normalize(pulls_json, ['@graph'])
df_col_pub.head()
```

```python
df_col_pub.info()
```

```python
df_col_pub_2 = df_col_pub[["title", "address.street-address", "location.latitude", "location.longitude", 
                 "organization.organization-name"]]
df_col_pub_2.head()
```

```python
df_col_pub_2["Type of place"] = df_col_pub_2["organization.organization-name"].str.split(" ", expand = True)[0] + " " + df_col_pub_2["organization.organization-name"].str.split(" ", expand = True)[1]
df_col_pub_2.head()
```

```python
df_col_pub_3 = df_col_pub_2[["title", "address.street-address", "location.latitude", 
                             "location.longitude", "Type of place"]]
df_col_pub_3.head()
```

```python
df_col_pub_3.columns = [["Place of interest", "Place address", "Latitude", "Longitude", "Type of place"]]
df_col_pub_3.head()
```
